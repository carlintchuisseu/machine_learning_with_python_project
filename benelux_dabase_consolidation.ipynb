```python
import pandas as pd
```

```python
df  = pd.read_excel("Bel-first_Export_1.xls",sheet_name = "Résultats")
for i in range(2,10):
    name = "Bel-first_Export_"+str(i)+".xls"
    df_read = pd.read_excel(name,sheet_name = "Résultats")
    df = pd.concat([df,df_read],ignore_index = True)
    print(df.shape)
    #print(name)
    
```

```python
df2  = pd.read_excel("Bel-first_Export_10.xls",sheet_name = "Résultats")
for i in range(11,21):
    name = "Bel-first_Export_"+str(i)+".xls"
    df_read = pd.read_excel(name,sheet_name = "Résultats")
    df2 = pd.concat([df2,df_read],ignore_index = True)
    print(df.shape)
    #print(name)
```

```python
df3  = pd.read_excel("Bel-first_Export_21.xls",sheet_name = "Résultats")
for i in range(22,33):
    name = "Bel-first_Export_"+str(i)+".xls"
    df_read = pd.read_excel(name,sheet_name = "Résultats")
    df3 = pd.concat([df3,df_read],ignore_index = True)
    print(df3.shape)
    #print(name)
```

```python
df_total = pd.concat([df,df2,df3],ignore_index = True)
df_total.shape
```
